```python
import pandas as pd
df = pd.read_csv('SMSSpamCollection.txt',sep='\t',header=None)
df.to_csv('dataset.csv')
df.head()
```

```python
df.rename(columns={0:'Spam/ham',1:'text'},inplace=True)
```

```python
df.head()
```

```python
df['Spam/ham']=df['Spam/ham'].map({'ham':0,'spam':1})
df['Spam/ham'].value_counts()
```

```python
df.isnull().sum()
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
import contractions
text = []
stop_words = stopwords.words('english')
for i in df['text']:
    temp = word_tokenize(str(i))
    temp = [contractions.fix(lemmatizer.lemmatize(i.lower())) for i in temp if(i not in stop_words and i.isalnum() == True) and len(i)>2]
    str1 = " " 
    print(str1.join(temp))
    text.append(str1.join(temp))
df['text']=text
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
c = CountVectorizer()
features = c.fit_transform(df['text'])
target = df['Spam/ham']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(features,target,test_size=0.2)

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
dt_clf = DecisionTreeClassifier()
dt_clf.fit(X_train,y_train)
accuracy_score(dt_clf.predict(X_test),y_test)
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
sss = StratifiedShuffleSplit(n_splits=1,test_size=0.2)
sss.get_n_splits(target)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn import svm
from sklearn.svm import SVC
classifiers = [
    SVC(kernel="linear", C=0.025),
    KNeighborsClassifier(),
    OneVsRestClassifier(svm.LinearSVC()),
#     RandomForestClassifier(max_depth=5, n_estimators=10, max_features=10),
   ]
for i, j in sss.split(features,target):
    X_train, X_test = features[i], features[j]
    y_train, y_test = target[i], target[j]
for clf in classifiers:
    score=0
    clf.fit(X_train, y_train)
    score=score+clf.score(X_test, y_test)
    print(score)
```

```python
from sklearn.metrics import accuracy_score
print('Accuracy Score',accuracy_score(y_test,clf.predict(X_test)))  
classifier = clf.fit(X_train, y_train)
score=clf.score(X_test, y_test)
y_test.value_counts()
```
